```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# -----------------------------
# Input Data
# -----------------------------
X = np.array([1, 2, 3, 4]).reshape(-1, 1)
Y = np.array([2, 3, 5, 4])

# -----------------------------
# Create and Train Model
# -----------------------------
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# -----------------------------
# Regression Equation
# -----------------------------
intercept = model.intercept_
slope = model.coef_[0]

print("========= Linear Regression =========")
print(f"Intercept (b0): {intercept:.4f}")
print(f"Slope (b1): {slope:.4f}")
print(f"\nRegression Equation:")
print(f"Y = {intercept:.4f} + {slope:.4f}X")

# -----------------------------
# Performance Metrics
# -----------------------------
mae = mean_absolute_error(Y, Y_pred)
mse = mean_squared_error(Y, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y, Y_pred)

print("\n========= Performance Metrics =========")
print(f"MAE  : {mae:.4f}")
print(f"MSE  : {mse:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R² Score : {r2:.4f}")

# -----------------------------
# Actual vs Predicted Values
# -----------------------------
print("\nActual\tPredicted")
for actual, pred in zip(Y, Y_pred):
    print(f"{actual}\t{pred:.2f}")

# ==========================================================
# Graph 1: Scatter Plot with Regression Line
# ==========================================================
plt.figure(figsize=(8,5))

plt.scatter(X, Y, color='blue', s=80, label='Actual Data')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')

plt.title("Linear Regression")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (Y)")
plt.legend()
plt.grid(True)

plt.show()

# ==========================================================
# Graph 2: Actual vs Predicted
# ==========================================================
plt.figure(figsize=(8,5))

plt.plot(X, Y, 'bo-', linewidth=2, label='Actual')
plt.plot(X, Y_pred, 'rs--', linewidth=2, label='Predicted')

plt.title("Actual vs Predicted Values")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (Y)")
plt.legend()
plt.grid(True)

plt.show()

# ==========================================================
# Graph 3: Residual Plot
# ==========================================================
residuals = Y - Y_pred

plt.figure(figsize=(8,5))

plt.scatter(Y_pred, residuals, color='green', s=80)
plt.axhline(y=0, color='red', linestyle='--')

plt.title("Residual Plot")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.grid(True)

plt.show()
```
